## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
#attributes or columns 
columns=['Clothing','Equipment','Accessories']
# stores 
store=['Delhi','Mumbai','Chennai','Kolkata','Hyderabad']
#clothing  
c = np.array([82,78,48,46,39])
# Equipment
e = np.array([48,31,51,29,18])
#Accessories
a = np.array([15,20,14,13,9])
ind=[x for x,_ in enumerate(store)]

# marking x and y label
plt.xlabel('Stores')
plt.ylabel('Range')

plt.bar(ind, c, width=0.5, color='b', bottom = e+a)
plt.bar(ind, e, width=0.5, color='r', bottom = a)
plt.bar(ind, a, width=0.5, color='g')

# xticks with an angle of 90 degree
plt.xticks(ind,store,rotation='90')

# legend 
plt.legend(columns,title="Stores")

plt.title('Stacked Bar Graph')
plt.show()
```

## Q.5 For Titanic Dataset
- Load Dataset using Seaborn
- Plot a Bar Graph of Survival of Passenger against the Passenger Class.
- Use proper "hue" attribute, and palette color. Mark 'x' and 'y' label with proper title.
- Use legend located at upper left corner

```python
data=ss.load_dataset('titanic')
data.head(5)
```

```python
n=list(data.pclass.unique())
count1=[data[(data['survived']==0) & (data['pclass']==1)].pclass.count(),data[(data['survived']==0) & (data['pclass']==2)].pclass.count(),data[(data['survived']==0) & (data['pclass']==3)].pclass.count()]
count2=[data[(data['survived']==1) & (data['pclass']==1)].pclass.count(),data[(data['survived']==1) & (data['pclass']==2)].pclass.count(),data[(data['survived']==1) & (data['pclass']==3)].pclass.count()]
x=np.arange(len(n))
plt.bar(x-.2,count1,width=0.40,color='black',label='0')
plt.bar(x+.2,count2,width=0.40,color='blue',label='1')
plt.legend(title='Survived',loc='upper left')
plt.xticks(x,n)
plt.xlabel("pclass")
plt.ylabel("passengers count")
plt.title("Bar Graph of Survival - Passenger Class",color='black',size=12)
plt.show()
```

## Q.9 


- Plot the Pie Chart for the two colums with some explode for the country "India"  and proper shadow

```python
country=['United States','Great Britain','India','Russia','Germany']
medal=[46,27,26,19,17]
ex=[0,0,0.2,0,0]
color=['c','yellowgreen','lightcoral','lightskyblue','gold']
# Plot
plt.pie(medal, explode=ex, labels=country, colors=color, shadow=True, pctdistance=0.6,startangle=150, autopct='%1.2f%%')
plt.axis('equal')
plt.title('Percentage distribution of medals among countries',color='b',size=15)
plt.show()
```

## Q.10 For Tips Dataset
- Load the dataset using seaborn
- Draw the Strip Plot between 'total bill' and 'day'
- Rotate xticks with an angle of 45 Degree

```python
# load the dataset
data1=ss.load_dataset('tips')
```

```python
# strip plot graph
x=ss.stripplot(x='day',y='total_bill',data=data1,size=5)
plt.xticks(rotation=45)
x.set_xlabel('day',fontsize=15)
x.set_ylabel('total_bill',fontsize=15)
plt.show()
```
